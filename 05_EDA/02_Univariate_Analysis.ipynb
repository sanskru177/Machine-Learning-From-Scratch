## 📖 Introduction

Univariate Analysis is the process of analyzing a single variable independently to understand its distribution, central tendency, spread, and frequency.

It is the first step in Exploratory Data Analysis (EDA).

## ❓ What is Univariate Analysis?

Univariate Analysis focuses on only one feature at a time.

Examples:

- Age
- Fare
- Sex
- Survived

No relationship with other variables is considered.

## 📚 Types of Variables

### Numerical Variables
- Age
- Fare
- SibSp
- Parch

### Categorical Variables
- Sex
- Embarked
- Survived
- Pclass

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Sanskruti\sanskruti document\Machine-Learning-From-Scratch\datasets\train.csv")
```

```python
df["Age"].describe()
```

```python
df["Age"].mean()
```

```python
df["Age"].median()
```

```python
df["Age"].mode()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Age"].dropna(), bins=20)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(df["Age"].dropna())
plt.title("Box Plot of Age")
plt.show()
```

## 📌 Analyze Categorical Feature (Sex)

```python
df["Sex"].value_counts()
```

```python
df["Sex"].value_counts(normalize=True) * 100
```

```python
df["Sex"].value_counts().plot(kind="bar")

plt.title("Gender Distribution")

plt.xlabel("Gender")

plt.ylabel("Count")

plt.show()
```

```python
df["Sex"].value_counts().plot(kind="pie", autopct="%1.1f%%")

plt.ylabel("")

plt.title("Gender Distribution")

plt.show()
```

## 📌 Analyze Target Variable

```python
df["Survived"].value_counts()
```

```python
df["Survived"].value_counts().plot(kind="bar")

plt.title("Survival Count")

plt.show()
```

## 🔍 Observations

- Most passengers were between 20 and 40 years old.
- The Age distribution is slightly right-skewed.
- The Sex column contains more males than females.
- More passengers did not survive than survived.

## ✅ Best Practices

- Analyze each feature independently before studying relationships.
- Use histograms for numerical variables.
- Use bar charts for categorical variables.
- Compare mean and median to understand skewness.
- Always investigate outliers before removing them.
