## 📖 Introduction

Data Visualization is the graphical representation of data. It helps transform raw numbers into meaningful insights, making it easier to identify trends, patterns, relationships, and outliers.

Visualization is one of the most important skills for Data Analysts and Data Scientists because it helps communicate findings effectively.

## 💡 Why is Data Visualization Important?

Data Visualization helps to:

- Understand complex datasets.
- Identify trends and patterns.
- Detect outliers.
- Compare categories.
- Present results clearly.
- Support business decision-making.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Sanskruti\sanskruti document\Machine-Learning-From-Scratch\datasets\train.csv")

df.head()
```

## 📌 1. Line Plot
Best Used For
Trends over time
Continuous data

```python
plt.figure(figsize=(8,5))

plt.plot(df.index, df["Fare"])

plt.title("Fare by Passenger Index")

plt.xlabel("Passenger Index")

plt.ylabel("Fare")

plt.show()
```

## 📌 2. Bar Chart
Best Used For
Comparing categories

```python
df["Pclass"].value_counts().plot(kind="bar")

plt.title("Passengers in Each Class")

plt.xlabel("Passenger Class")

plt.ylabel("Count")

plt.show()
```

## 📌 3. Histogram
Best Used For
Distribution of numerical data

```python
plt.figure(figsize=(8,5))

plt.hist(df["Age"].dropna(), bins=20)

plt.title("Age Distribution")

plt.xlabel("Age")

plt.ylabel("Frequency")

plt.show()
```

## 📌 4. Scatter Plot
Best Used For
Relationship between two numerical variables

```python
plt.figure(figsize=(8,5))

plt.scatter(df["Age"], df["Fare"])

plt.title("Age vs Fare")

plt.xlabel("Age")

plt.ylabel("Fare")

plt.show()
```

## 📌 5. Box Plot
Best Used For
Detecting outliers

```python
plt.figure(figsize=(6,4))

plt.boxplot(df["Fare"].dropna())

plt.title("Box Plot of Fare")

plt.ylabel("Fare")

plt.show()
```

## 📌 6. Pie Chart
Best Used For
Showing proportions

```python
df["Sex"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Gender Distribution")

plt.ylabel("")

plt.show()
```

## ✅ Best Practices

- Select the chart based on the type of data.
- Always include titles and axis labels.
- Avoid unnecessary colors and decorations.
- Keep visualizations simple and readable.
- Interpret the chart instead of only creating it.
