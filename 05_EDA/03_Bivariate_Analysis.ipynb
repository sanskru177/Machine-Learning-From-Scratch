## 📖 Introduction

Bivariate Analysis studies the relationship between two variables. It helps us understand whether one variable influences another and uncovers trends, patterns, and associations in the data.

Unlike Univariate Analysis, which focuses on a single variable, Bivariate Analysis compares two variables simultaneously.

## ❓ What is Bivariate Analysis?

Bivariate Analysis examines the relationship between two variables.

Examples:

- Age vs Survived
- Sex vs Survived
- Fare vs Pclass
- Embarked vs Survived

It helps answer questions like:
- Does gender affect survival?
- Does passenger class influence fare?
- Is age related to survival?

## 📚 Types of Relationships

### Numerical vs Numerical
Example:
- Age vs Fare

### Numerical vs Categorical
Example:
- Age vs Survived

### Categorical vs Categorical
Example:
- Sex vs Survived

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Sanskruti\sanskruti document\Machine-Learning-From-Scratch\datasets\train.csv")
```

## 📌 1. Categorical vs Categorical

```python
pd.crosstab(df["Sex"], df["Survived"])
```

```python
pd.crosstab(df["Sex"], df["Survived"]).plot(kind="bar")

plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

## 📌 2. Passenger Class vs Survival

```python
pd.crosstab(df["Pclass"], df["Survived"])
```

```python
pd.crosstab(df["Pclass"], df["Survived"]).plot(kind="bar")

plt.title("Passenger Class vs Survival")

plt.xlabel("Passenger Class")

plt.ylabel("Count")

plt.show()
```

## 📌 3. Numerical vs Categorical

```python
df.groupby("Survived")["Age"].mean()
```

## 📌 4. Fare vs Passenger Class

```python
df.groupby("Pclass")["Fare"].mean()
```

```python
df.groupby("Pclass")["Fare"].mean().plot(kind="bar")

plt.title("Average Fare by Passenger Class")

plt.xlabel("Passenger Class")

plt.ylabel("Average Fare")

plt.show()
```

## 📌 5. Numerical vs Numerical

```python
plt.figure(figsize=(8,5))

plt.scatter(df["Age"], df["Fare"])

plt.title("Age vs Fare")

plt.xlabel("Age")

plt.ylabel("Fare")

plt.show()
```

```python
df[["Age","Fare"]].corr()
```

Correlation values:

+1 → Strong Positive
0 → No Relationship
-1 → Strong Negative

## 🔍 Observations

- Female passengers had a higher survival rate than males.
- First-class passengers survived more frequently.
- First-class passengers paid higher fares.
- The relationship between Age and Fare is weak.

## ✅ Best Practices

- Use Crosstab for two categorical variables.
- Use GroupBy to summarize numerical data by category.
- Use Scatter Plots for numerical variables.
- Use Correlation to measure linear relationships.
- Validate findings before drawing conclusions.
