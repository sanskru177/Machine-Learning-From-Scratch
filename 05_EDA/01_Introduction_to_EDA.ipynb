## 📖 Introduction

Exploratory Data Analysis (EDA) is the process of examining, summarizing, and visualizing data before applying machine learning algorithms.

The main goal of EDA is to understand the dataset, identify patterns, detect anomalies, and prepare the data for further analysis.

## ❓ What is Exploratory Data Analysis?

EDA is the first analytical step after data collection and cleaning.

It helps answer questions such as:

- What does the dataset contain?
- Are there missing values?
- Are there duplicate records?
- What are the distributions of features?
- Are there relationships between variables?
- Are there outliers?

## 💡 Why Perform EDA?

EDA helps to:

- Understand the structure of the dataset.
- Detect missing values and duplicates.
- Identify outliers.
- Discover trends and patterns.
- Select useful features.
- Improve machine learning model performance.

## 🔄 Typical EDA Workflow

1. Load the dataset.
2. Understand the data.
3. Inspect data types.
4. Check missing values.
5. Analyze numerical features.
6. Analyze categorical features.
7. Identify relationships between variables.
8. Visualize the data.
9. Draw conclusions.

- Understand feature meanings.
- Check data types.
