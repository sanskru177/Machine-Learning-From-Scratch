# 📖 Introduction

Every programming language follows a set of grammatical rules known as **syntax**. Just as incorrect grammar can make a sentence difficult to understand, incorrect syntax can prevent a program from running.

Python is known for its clean and readable syntax, making it one of the easiest programming languages to learn. Understanding Python syntax helps you write error-free code, improve readability, and build a strong programming foundation.

In addition to syntax, Python provides **keywords**—special reserved words that define the structure and behavior of a program. These keywords cannot be used as variable names, function names, or identifiers.

# 🤔 What is Python Syntax?

Python syntax is the set of rules that defines how Python programs should be written. These rules determine how statements, expressions, variables, functions, loops, and other programming constructs are structured.

If the syntax rules are not followed, Python raises a **SyntaxError**, and the program will not execute.

### Example of Correct Syntax

```python
name = "Sanskruti"
print(name)
```

### Example of Incorrect Syntax

```python
if True
    print("Hello")
```

The above code generates a **SyntaxError** because the colon (`:`) is missing after the `if` statement.

# 📜 Rules of Python Syntax

Python follows several important syntax rules.

## 1️⃣ Indentation

Python uses indentation instead of braces to define code blocks.


## 2️⃣ Colon (:)

Statements like **if**, **for**, **while**, **def**, and **class** must end with a colon.


## 3️⃣ Case Sensitive

Python treats uppercase and lowercase letters differently.

Example:

```
name
Name
NAME
```

These are three different identifiers.


## 4️⃣ Statements

Each statement usually occupies one line.


## 5️⃣ Comments

Comments begin with the **#** symbol.


## 6️⃣ Proper Naming

Variable names should follow Python naming conventions.

# 🆔 Identifiers

Identifiers are names given to variables, functions, classes, and objects.

## Rules for Identifiers

- Must start with a letter or underscore (_)
- Cannot start with a digit
- Cannot contain spaces
- Cannot use Python keywords
- Can contain letters, digits, and underscores

### Valid Identifiers

```
name
_age
student1
total_marks
```

### Invalid Identifiers

