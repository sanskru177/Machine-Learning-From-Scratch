## 📖 Introduction

A **DataFrame** is a two-dimensional labeled data structure consisting of rows and columns. It is the most commonly used data structure in Pandas for storing, analyzing, and manipulating structured data.

## 📊 What is a DataFrame?

A DataFrame is similar to an Excel spreadsheet or SQL table.

It contains:

- Rows
- Columns
- Index
- Different data types

Each column represents a Series.

Creating a DataFrame
From Dictionary

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [22, 25, 23],
    "City": ["Delhi", "Mumbai", "Pune"]
}

df = pd.DataFrame(data)

print(df)
```

From List of Lists

```python
data = [
    ["Alice",22,"Delhi"],
    ["Bob",25,"Mumbai"],
    ["Charlie",23,"Pune"]
]

df = pd.DataFrame(
    data,
    columns=["Name","Age","City"]
)

print(df)
```

From List of Dictionaries

```python
data = [
    {"Name":"Alice","Age":22},
    {"Name":"Bob","Age":25},
    {"Name":"Charlie","Age":23}
]

df = pd.DataFrame(data)

print(df)
```

DataFrame Attributes

```python
print(df.shape)

print(df.size)

print(df.ndim)

print(df.columns)

print(df.index)

print(df.dtypes)

print(df.values)
```

Accessing Columns

```python
print(df["Name"])

print(df["Age"])
```

Selecting Multiple Columns

Adding a New Column

```python
df["Salary"] = [50000,60000,55000]

print(df)
```

Updating a Column

```python
df["Age"] = df["Age"] + 1

print(df)
```

Deleting a Column

```python
df.drop("Salary",axis=1,inplace=True)

print(df)
```

Renaming Columns

```python
df.rename(
    columns={
        "Name":"Employee_Name"
    },
    inplace=True
)

print(df)
```

Basic Statistics

```python
print(df.describe())
```

Data Types

```python
print(df.dtypes)
```

Sorting

```python
print(df.sort_values(by="Age"))
```

| Method          | Purpose             |
| --------------- | ------------------- |
| `DataFrame()`   | Create DataFrame    |
| `shape`         | Rows & Columns      |
| `size`          | Total Elements      |
| `columns`       | Column Names        |
| `index`         | Row Labels          |
| `dtypes`        | Data Types          |
| `values`        | NumPy Array         |
| `describe()`    | Statistical Summary |
| `rename()`      | Rename Columns      |
| `drop()`        | Delete Columns      |
| `sort_values()` | Sort Data           |
