# 📂 Reading and Writing Data in Pandas

## 🎯 Learning Objectives

- Read data from different file formats.
- Write DataFrames to files.
- Understand commonly used parameters.
- Load only required data.
- Save processed data efficiently.

## 📖 Introduction

Reading and writing data is one of the most common tasks in Pandas. It allows you to import data from files such as CSV, Excel, and JSON, perform analysis, and export the processed data for reporting or further use.

Import Pandas

```python
import pandas as pd
```

Read CSV File

```python
df = pd.read_csv("students.csv")

print(df)
```

## 📊 Read Excel File

```python
df = pd.read_excel("students.xlsx")

print(df)
```

## 🌐 Read JSON File

```python
df = pd.read_json("students.json")

print(df)
```

View First Rows

```python
print(df.head())
```

Read Specific Columns

```python
df = pd.read_csv(
    "students.csv",
    usecols=["Name","Age"]
)

print(df)
```

Read Limited Rows

```python
df = pd.read_csv(
    "students.csv",
    nrows=5
)

print(df)
```

Skip Rows

```python
df = pd.read_csv(
    "students.csv",
    skiprows=2
)

print(df)
```

Set Index Column

```python
df = pd.read_csv(
    "students.csv",
    index_col="ID"
)

print(df)
```

## 💾 Write CSV File

```python
df.to_csv(
    "output.csv",
    index=False
)
```

## 📊 Write Excel File

```python
df.to_excel(
    "output.xlsx",
    index=False
)
```

## 🌐 Write JSON File

```python
df.to_json(
    "output.json"
)
```

| Parameter   | Purpose               |
| ----------- | --------------------- |
| `usecols`   | Read selected columns |
| `nrows`     | Read limited rows     |
| `skiprows`  | Skip rows             |
| `index_col` | Set index column      |
| `header`    | Specify header row    |
| `names`     | Custom column names   |
| `sep`       | Change delimiter      |
| `encoding`  | File encoding         |


| Function       | File Type |
| -------------- | --------- |
| `read_csv()`   | CSV       |
| `read_excel()` | Excel     |
| `read_json()`  | JSON      |
| `to_csv()`     | CSV       |
| `to_excel()`   | Excel     |
| `to_json()`    | JSON      |


## 📝 Summary

- Use `read_csv()` to load CSV files.
- Use `read_excel()` for Excel files.
- Use `read_json()` for JSON files.
- Export data using `to_csv()`, `to_excel()`, and `to_json()`.
- Parameters like `usecols`, `nrows`, and `index_col` help load data efficiently.
