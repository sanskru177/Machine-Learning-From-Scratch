## 📖 Introduction

Pandas is an open-source Python library used for data manipulation, analysis, and cleaning. It provides powerful and flexible data structures that make working with structured data simple and efficient.

## 🐼 What is Pandas?

Pandas is a Python library built on top of NumPy. It is designed to work with structured and tabular data using two primary data structures:

- Series (1D)
- DataFrame (2D)

It simplifies data loading, cleaning, transformation, analysis, and visualization.

## ❓ Why Pandas?

- Easy data manipulation
- Fast data analysis
- Handles missing values
- Supports multiple file formats
- Built on NumPy
- Powerful indexing and filtering
- Widely used in Data Science and Machine Learning

## ⭐ Features

- Fast and efficient
- Label-based indexing
- Automatic data alignment
- Missing value handling
- Grouping and aggregation
- Data merging and joining
- Time series support
- Integration with NumPy and Matplotlib

## 📦 Data Structures

### 1. Series
A one-dimensional labeled array.

### 2. DataFrame
A two-dimensional table consisting of rows and columns.

## ❌ Limitations

- Higher memory usage than NumPy
- Slower than NumPy for pure numerical computations
- Not suitable for distributed big data processing

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [22, 25, 23]
}

df = pd.DataFrame(data)

print(df)
```
