## 📖 Introduction

A **Series** is a one-dimensional labeled array capable of holding any data type such as integers, floats, strings, or Python objects. It is one of the two fundamental data structures in Pandas.

## 📊 What is a Series?

A Series consists of:

- Values
- Index

Each value is associated with an index label, making data retrieval fast and efficient.

```python
import pandas as pd

s = pd.Series([10, 20, 30, 40, 50])

print(s)
```

```python
student = {
    "Math":90,
    "Science":85,
    "English":88
}

s = pd.Series(student)

print(s)
```

```python
s = pd.Series(
    [90,85,88],
    index=["Math","Science","English"]
)

print(s)
```

```python
print(s[0])
print(s[1])
```

By Label

```python
print(s["Math"])
```

Slicing

```python
print(s[0:2])
```

```python
print(s.index)
print(s.values)
print(s.dtype)
print(s.shape)
print(s.size)
print(s.ndim)
```

Common Methods

```python
print(s.head())
print(s.tail())
print(s.describe())
print(s.max())
print(s.min())
print(s.mean())
print(s.sum())
```

Vectorized Operations

```python
print(s + 10)

print(s * 2)
```

Boolean Filtering

```python
print(s[s > 85])
```

Checking Missing Values

```python
print(s.isna())
```

| Method       | Purpose        |
| ------------ | -------------- |
| `Series()`   | Create Series  |
| `head()`     | First rows     |
| `tail()`     | Last rows      |
| `describe()` | Statistics     |
| `mean()`     | Average        |
| `sum()`      | Sum            |
| `max()`      | Maximum        |
| `min()`      | Minimum        |
| `isna()`     | Missing values |
| `values`     | Data           |
| `index`      | Labels         |


## 📝 Summary

- Series is a one-dimensional labeled array.
- Every Series contains values and indexes.
- Supports fast indexing and slicing.
- Provides built-in statistical methods.
- Forms the foundation of Pandas DataFrames.
