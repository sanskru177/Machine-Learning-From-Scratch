# 🔍 Data Inspection in Pandas

## 🎯 Learning Objectives

- Understand how to inspect a dataset.
- View rows, columns, and data types.
- Generate statistical summaries.
- Check memory usage and missing values.
- Explore the structure of a DataFrame.

## 📖 Introduction

Before cleaning or analyzing data, it is important to inspect the dataset. Data inspection helps understand the size, structure, data types, missing values, and overall quality of the data. It is one of the first steps in every Data Analysis and Machine Learning project.

## 📌 Import Pandas

Import the Pandas library using the standard alias `pd`.

```python
import pandas as pd
```

````python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [22, 25, 23, 24, 21],
    "Marks": [85, 90, 78, 88, 92]
}

df = pd.DataFrame(data)

print(df)
````

## 📌 `head()`

Displays the first **5 rows** by default. You can also specify the number of rows.

```python
df.head()

df.head(3)
```

## 📌 `tail()`

Displays the last **5 rows** by default.

```python
df.tail()

df.tail(2)
```

## 📌 `info()`

Displays information about the DataFrame including:
- Number of rows and columns
- Column names
- Data types
- Non-null values
- Memory usage

```python
df.info()
```

## 📌 `describe()`

Generates descriptive statistics for numerical columns.

```python
df.describe()
```

## 📌 `shape`

Returns the number of rows and columns.

```python
print(df.shape)
```

## 📌 `size`

Returns the total number of elements.

```python
print(df.size)
```

## 📌 `columns`

Returns the column names.

```python
print(df.columns)
```

## 📌 `index`

Returns the row labels (index).

```python
print(df.index)
```

## 📌 `dtypes`

Displays the data type of each column.

```python
print(df.dtypes)
```

## 📌 `sample()`

Returns random rows from the DataFrame.

```python
df.sample()

df.sample(2)
```

## 📌 `memory_usage()`

Shows the memory consumed by each column.

```python
df.memory_usage()
```

## 📌 `isnull()`

Checks whether values are missing.

```python
df.isnull()

df.isnull().sum()
```

| Method           | Purpose                    |
| ---------------- | -------------------------- |
| `head()`         | View first rows            |
| `tail()`         | View last rows             |
| `info()`         | Dataset information        |
| `describe()`     | Statistical summary        |
| `shape`          | Number of rows and columns |
| `size`           | Total elements             |
| `columns`        | Column names               |
| `index`          | Row labels                 |
| `dtypes`         | Data types                 |
| `sample()`       | Random rows                |
| `memory_usage()` | Memory consumption         |
| `isnull()`       | Check missing values       |


## 📝 Summary

- Data inspection is the first step in data analysis.
- `head()` and `tail()` provide a quick preview of the dataset.
- `info()` gives an overview of the DataFrame structure.
- `describe()` summarizes numerical data.
- `shape`, `columns`, and `dtypes` help understand the dataset.
- `memory_usage()` and `isnull()` are useful for optimization and data cleaning.
