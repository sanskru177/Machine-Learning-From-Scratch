
# 🎯 Selecting and Indexing Data in Pandas

## 🎯 Learning Objectives

- Select rows and columns from a DataFrame.
- Understand label-based and position-based indexing.
- Use `loc[]` and `iloc[]` effectively.
- Apply Boolean Indexing.
- Set and reset indexes.

## 📖 Introduction

Selecting and indexing data is one of the most common tasks in Pandas. It allows you to retrieve specific rows, columns, or subsets of data efficiently for analysis, cleaning, and machine learning.

## 📌 Import Pandas

```python
import pandas as pd
```

````python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [22, 25, 23, 24],
    "Marks": [85, 90, 78, 88],
    "City": ["Delhi", "Mumbai", "Pune", "Hyderabad"]
}

df = pd.DataFrame(data)

print(df)
````

## 📌 Select a Single Column

Returns a Pandas Series.

```python
df["Name"]
```

````python
print(df['Name'])
````

## 📌 Select Multiple Columns

Returns a DataFrame.

```python
df[["Name", "Marks"]]
```

````python
print(df[["Name", "Marks"]])
````

## 📌 `loc[]` (Label-Based Indexing)

Select rows and columns using index labels.

```python
df.loc[0]

df.loc[1:3]

df.loc[:, ["Name", "Marks"]]

df.loc[0:2, ["Name", "Age"]]
```

````python
df.loc[0]

df.loc[1:3]

df.loc[:, ["Name", "Marks"]]

df.loc[0:2, ["Name", "Age"]]
````

## 📌 `iloc[]` (Position-Based Indexing)

Select rows and columns using integer positions.

```python
df.iloc[0]

df.iloc[0:3]

df.iloc[:, 1:3]

df.iloc[1:4, 0:2]
```

````python
df.iloc[0]
````

````python
df.iloc[0:3]
````

## 📌 Boolean Indexing

Filter data based on conditions.

```python
df[df["Marks"] > 85]

df[df["Age"] < 24]
```

````python
df[df["Marks"] > 85]
````

## 📌 Multiple Conditions

Use `&` (AND) and `|` (OR).

```python
df[(df["Age"] > 22) & (df["Marks"] > 80)]

df[(df["City"] == "Pune") | (df["City"] == "Delhi")]
```

````python
df[(df["Age"] > 22) & (df["Marks"] > 80)]

df[(df["City"] == "Pune") | (df["City"] == "Delhi")]
````

## 📌 `set_index()`

Set a column as the DataFrame index.

```python
df.set_index("Name", inplace=True)

print(df)
```

## 📌 `reset_index()`

Restore the default integer index.

```python
df.reset_index(inplace=True)

print(df)
```

## 📌 Access Specific Values

```python
df.at[0, "Name"]

df.iat[0, 1]
```

| Method           | Purpose                                   |
| ---------------- | ----------------------------------------- |
| `df["Column"]`   | Select single column                      |
| `df[[...]]`      | Select multiple columns                   |
| `loc[]`          | Label-based indexing                      |
| `iloc[]`         | Position-based indexing                   |
| `at[]`           | Fast access to a single value by label    |
| `iat[]`          | Fast access to a single value by position |
| `set_index()`    | Set index                                 |
| `reset_index()`  | Reset index                               |
| Boolean Indexing | Filter rows                               |


## 📝 Summary

- Use `[]` to select columns.
- `loc[]` uses labels, while `iloc[]` uses integer positions.
- Boolean indexing filters data based on conditions.
- `set_index()` changes the index column.
- `reset_index()` restores the default index.
- `at[]` and `iat[]` provide fast access to individual values.
