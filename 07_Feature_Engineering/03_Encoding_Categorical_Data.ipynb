## 📖 Introduction

Machine Learning algorithms work with numerical data, but real-world datasets often contain categorical features such as Gender, City, Education, or Country.

Encoding converts these categorical values into numerical values so that machine learning models can process them.

## ❓ What is Categorical Data?

Categorical data represents categories or labels instead of numerical values.

Examples:

- Male
- Female
- Red
- Blue
- Mumbai
- Pune

These values must be converted into numbers before training most machine learning models.

## 📚 Types of Categorical Data

### Nominal Data

No natural order exists.

Examples:

- Gender
- City
- Country
- Color

---

### Ordinal Data

Categories have a meaningful order.

Examples:

Education:

High School < Bachelor's < Master's < PhD

Customer Satisfaction:

Low < Medium < High

## 💡 Why Encoding?

Machine Learning algorithms perform mathematical operations.

Algorithms cannot directly understand text such as:

Male

Female

Mumbai

Delhi

Encoding converts these values into numbers while preserving useful information.

## 📌 Label Encoding

Label Encoding assigns a unique integer to each category.

Example:

Male → 1

Female → 0

Best used for:

Binary categorical variables.

```python
le = LabelEncoder()

df["Sex_Label"] = le.fit_transform(df["Sex"])

df[["Sex", "Sex_Label"]].head()
```

## 📌 One-Hot Encoding

One-Hot Encoding creates a new column for each category.

Example:

Embarked

C

Q

S

becomes

Embarked_C

Embarked_Q

Embarked_S

Best for Nominal Data.

```python
embarked_encoded = pd.get_dummies(df["Embarked"], prefix="Embarked")

embarked_encoded.head()
```

```python
df = pd.concat([df, embarked_encoded], axis=1)

df.head()
```

## 📌 Ordinal Encoding

Ordinal Encoding assigns numbers based on the natural order of categories.

Example:

Low → 0

Medium → 1

High → 2

Use only for Ordinal Data.

```python
education = pd.DataFrame({
    "Education": [
        "High School",
        "Bachelor",
        "Master",
        "PhD",
        "Bachelor"
    ]
})

education
```

```python
oe = OrdinalEncoder(
    categories=[[
        "High School",
        "Bachelor",
        "Master",
        "PhD"
    ]]
)

education["Education_Encoded"] = oe.fit_transform(
    education[["Education"]]
)

education
```

## 📌 Frequency Encoding

Frequency Encoding replaces each category with its frequency.

Example:

City

Mumbai → 250

Pune → 180

Delhi → 150

Useful when there are many categories.

```python
frequency = df["Embarked"].value_counts()

frequency
```

```python
df["Embarked_Frequency"] = df["Embarked"].map(frequency)

df[["Embarked","Embarked_Frequency"]].head()
```

## 📌 Target Encoding

Target Encoding replaces each category with the average value of the target variable.

Example:

Male

Average Survival = 0.18

Female

Average Survival = 0.74

Useful for high-cardinality categorical variables.

⚠️ Be careful to avoid data leakage.

```python
target_encoding = df.groupby("Sex")["Survived"].mean()

target_encoding
```

```python
df["Sex_Target"] = df["Sex"].map(target_encoding)

df[["Sex","Sex_Target"]].head()
```

## 📌 Binary Encoding

Binary Encoding first assigns an integer to each category and then converts that integer into binary digits.

It is useful when a feature has many unique categories and One-Hot Encoding would create too many columns.

## 📌 Comparison of Encoding Techniques

| Encoding           | Best For              | Advantages                 | Disadvantages          |
| ------------------ | --------------------- | -------------------------- | ---------------------- |
| Label Encoding     | Binary categories     | Simple and fast            | May imply order        |
| One-Hot Encoding   | Nominal data          | No false ordering          | Creates many columns   |
| Ordinal Encoding   | Ordered categories    | Preserves order            | Only for ordinal data  |
| Frequency Encoding | High-cardinality data | Reduces dimensions         | Loses category meaning |
| Target Encoding    | High-cardinality data | Often improves performance | Risk of data leakage   |
| Binary Encoding    | Many categories       | Memory efficient           | Harder to interpret    |

## 🔍 Observations

- Machine learning algorithms require numerical input.
- Different encoding techniques are suitable for different types of categorical data.
- Choosing the right encoding method can improve model performance.

## ✅ Best Practices

- Use Label Encoding for binary variables.
- Use One-Hot Encoding for nominal categories.
- Use Ordinal Encoding only when categories have a natural order.
- Avoid high-dimensional One-Hot Encoding for features with many unique values.
- Be cautious of data leakage when using Target Encoding.
