## 📖 Introduction

Feature Engineering is one of the most important steps in the Machine Learning pipeline.

A machine learning model learns patterns from the input features (columns). The better the quality of these features, the better the model's performance.

Feature Engineering is the process of creating, modifying, selecting, or transforming features to improve the accuracy and performance of machine learning models.

Many real-world machine learning projects spend more time on feature engineering than on model building.

## ❓ What is a Feature?

A Feature is an individual measurable property or characteristic used as input for a machine learning model.

Examples:

Age

Salary

Gender

Height

Weight

Marks

Experience

Each column in a dataset is usually considered a feature.

## 📚 Feature vs Target Variable

Feature (Independent Variable)

- Input provided to the model.
- Used to make predictions.

Examples:

Age

Fare

Pclass

Sex

---

Target Variable (Dependent Variable)

- Output the model predicts.

Example:

Titanic Dataset

Target Variable = Survived

## 💡 Why is Feature Engineering Important?

Good feature engineering can:

- Improve model accuracy.
- Reduce overfitting.
- Reduce training time.
- Improve model interpretability.
- Help algorithms learn patterns more effectively.

Bad features often lead to poor model performance, even if a powerful algorithm is used.

## 📚 Types of Feature Engineering

Common Feature Engineering techniques include:

- Handling Missing Values
- Encoding Categorical Variables
- Feature Scaling
- Feature Transformation
- Feature Creation
- Feature Selection
- Handling Imbalanced Data
- Dimensionality Reduction

## 📌 Feature Engineering in Titanic Dataset

| Original Feature | Possible Engineered Feature |
| ---------------- | --------------------------- |
| Age              | Age Group                   |
| Fare             | Fare Category               |
| Name             | Title (Mr, Mrs, Miss, etc.) |
| Cabin            | Deck                        |
| Ticket           | Ticket Prefix               |
| SibSp + Parch    | Family Size                 |


```python
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1

df[["SibSp", "Parch", "FamilySize"]].head()
```

Explanation:

FamilySize represents the total number of family members traveling together, including the passenger.

```python
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[0, 12, 18, 35, 60, 100],
    labels=["Child", "Teen", "Young Adult", "Adult", "Senior"]
)

df[["Age", "AgeGroup"]].head()
```

Observation:

Continuous numerical data has been converted into meaningful categories.

## ✅ Best Practices

- Understand the dataset before creating new features.
- Avoid creating unnecessary features.
- Handle missing values before feature engineering.
- Document every engineered feature.
- Validate whether new features actually improve model performance.
