# 📘 Introduction

Not all Machine Learning models learn in the same way. Some models learn using labeled data, while others discover patterns from unlabeled data. Certain models learn from a combination of both, and others improve by interacting with an environment through rewards and penalties.

Based on the learning approach, Machine Learning is divided into four main categories:

- 📌 Supervised Learning
- 📌 Unsupervised Learning
- 📌 Semi-Supervised Learning
- 📌 Reinforcement Learning

Each type is designed to solve different kinds of real-world problems.

# 🎓 Supervised Learning

Supervised Learning is a type of Machine Learning where the model is trained using **labeled data**. This means that each input data point has a corresponding correct output (target). The model learns the relationship between the input and output so that it can predict the correct output for new, unseen data.

> **Key Idea:** The model learns from examples where the correct answers are already known.



## 📊 Types of Supervised Learning

### 1️⃣ Regression

Regression is used when the output is a **continuous numerical value**.

**Examples:**
- Predicting house prices
- Predicting stock prices
- Predicting temperature


### 2️⃣ Classification

Classification is used when the output belongs to a **specific category**.

**Examples:**
- Spam or Not Spam
- Disease or No Disease
- Fraud or Not Fraud
- Cat or Dog Classification

## 🤖 Popular Supervised Learning Algorithms

| Algorithm | Used For |
|-----------|----------|
| Linear Regression | Predicting numerical values |
| Logistic Regression | Binary classification |
| Decision Tree | Classification & Regression |
| Random Forest | Classification & Regression |
| Support Vector Machine (SVM) | Classification |
| K-Nearest Neighbors (KNN) | Classification & Regression |
| Naive Bayes | Text Classification |
