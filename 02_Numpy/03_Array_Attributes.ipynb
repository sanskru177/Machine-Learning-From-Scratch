# 📊 NumPy Array Attributes

## 🎯 Learning Objectives

- Understand NumPy array attributes.
- Learn how to inspect an array.
- Differentiate between shape, size, and dimension.
- Understand memory usage of arrays.
- Convert array data types.

## 📖 Introduction

Every NumPy array contains metadata that describes its structure, size, data type, and memory usage. These properties are known as **Array Attributes**. Understanding these attributes is essential for data manipulation, debugging, and optimizing machine learning workflows.

````python
import numpy as np

arr = np.array([[10, 20, 30],
                [40, 50, 60]])

print(arr)
````

## 🔹 `ndim`

Returns the **number of dimensions (axes)** of the array.

- 1D → One axis
- 2D → Two axes
- 3D → Three axes

````python
print(arr.ndim)
````

## 🔹 `shape`

Returns the shape of the array as **(rows, columns)**.

Syntax:

```python
array.shape
```

````python
print(arr.shape)
````

## 🔹 `size`

Returns the **total number of elements** in the array.

````python
print(arr.size)
````

## 🔹 `dtype`

Returns the **data type** of array elements.

````python
print(arr.dtype)
````

## 🔹 `itemsize`

Returns the size of **one array element** in bytes.

````python
print(arr.itemsize)
````

## 🔹 `nbytes`

Returns the **total memory occupied** by the array.

Formula:

Total Memory = size × itemsize

````python
print(arr.nbytes)
````

## 🔹 `T`

Returns the transpose of an array by interchanging rows and columns.

````python
print(arr.T)
````

## 🔹 `astype()`

Converts the array from one data type to another.

````python
new_arr = arr.astype(float)
print(new_arr)
print(new_arr.dtype)
````

## 📝 Summary

- `ndim` → Number of dimensions
- `shape` → Rows and columns
- `size` → Total elements
- `dtype` → Element data type
- `itemsize` → Bytes per element
- `nbytes` → Total memory used
- `T` → Transpose the array
- `astype()` → Convert data type
