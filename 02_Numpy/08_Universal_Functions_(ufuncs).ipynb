# ⚡ NumPy Universal Functions (ufuncs)

## 🎯 Learning Objectives

- Understand Universal Functions (ufuncs).
- Learn unary and binary ufuncs.
- Perform mathematical operations efficiently.
- Use trigonometric, exponential, and logarithmic functions.
- Apply rounding functions.

```python
import numpy as np

arr = np.array([1, 4, 9, 16, 25])
```

## 🔹 Unary Functions

Unary ufuncs operate on a single array.

```python
print(np.sqrt(arr))
```

```python
print(np.square(arr))
```

```python
arr = np.array([-10, -20, 30])

print(np.abs(arr))
```

```python
arr = np.array([1, 2, 3])

print(np.exp(arr))
```

## 🔹 Logarithmic Functions

```python
arr = np.array([1, 10, 100])

print(np.log(arr))
print(np.log10(arr))
print(np.log2(arr))
```

## 🔹 Trigonometric Functions

```python
angles = np.array([0, np.pi/2, np.pi])

print(np.sin(angles))
print(np.cos(angles))
print(np.tan(angles))
```

## 🔹 Rounding Functions

```python
arr = np.array([2.3, 4.8, 5.5])

print(np.round(arr))
print(np.floor(arr))
print(np.ceil(arr))
```

## 🔹 Binary Functions

Binary ufuncs operate on two arrays.

```python
a = np.array([10,20,30])
b = np.array([1,2,3])

print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

```python
print(np.maximum(a,b))
print(np.minimum(a,b))
```

## 📝 Summary

- Ufuncs perform fast element-wise operations.
- Unary ufuncs operate on one array.
- Binary ufuncs operate on two arrays.
- NumPy provides mathematical, logarithmic, trigonometric, and rounding functions.
- Ufuncs are optimized for high-performance numerical computing.
