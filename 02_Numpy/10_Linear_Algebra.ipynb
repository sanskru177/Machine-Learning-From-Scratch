# 🧮 NumPy Linear Algebra

## 🎯 Learning Objectives

- Understand vectors and matrices.
- Perform matrix operations.
- Learn matrix multiplication techniques.
- Compute determinant, inverse, rank, and trace.
- Calculate eigenvalues and eigenvectors.
- Solve systems of linear equations.

## 📖 Introduction

Linear Algebra is the mathematical foundation of Machine Learning. NumPy provides the `numpy.linalg` module, which offers efficient functions for matrix computations used in algorithms like Linear Regression, PCA, SVM, and Neural Networks.

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])
```

## ➕ Matrix Addition

```python
print(A + B)
```

## ➖ Matrix Subtraction

```python
print(A - B)
```

## ✖️ Matrix Multiplication

```python
print(np.dot(A, B))
```

```python
print(np.matmul(A, B))
```

```python
print(A @ B)
```

## 🔄 Transpose

```python
print(A.T)
```

## 📐 Determinant

```python
print(np.linalg.det(A))
```

## 🔁 Inverse

```python
print(np.linalg.inv(A))
```

## 📊 Rank

```python
print(np.linalg.matrix_rank(A))
```

## 📈 Trace

```python
print(np.trace(A))
```

## 🧩 Eigenvalues & Eigenvectors

```python
values, vectors = np.linalg.eig(A)

print("Eigenvalues:", values)
print("Eigenvectors:\n", vectors)
```

## 📝 Solving Linear Equations

```python
A = np.array([[2, 1],
              [1, 3]])

B = np.array([8, 13])

X = np.linalg.solve(A, B)

print(X)
```

| Function                  | Purpose                    |
| ------------------------- | -------------------------- |
| `np.dot()`                | Matrix multiplication      |
| `np.matmul()`             | Matrix multiplication      |
| `@`                       | Matrix multiplication      |
| `A.T`                     | Transpose                  |
| `np.linalg.det()`         | Determinant                |
| `np.linalg.inv()`         | Inverse                    |
| `np.linalg.matrix_rank()` | Rank                       |
| `np.trace()`              | Trace                      |
| `np.linalg.eig()`         | Eigenvalues & Eigenvectors |
| `np.linalg.solve()`       | Solve linear equations     |
