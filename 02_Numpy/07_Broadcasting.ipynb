## 📖 Introduction

Broadcasting is a powerful NumPy feature that allows arithmetic operations on arrays with different shapes without explicitly copying data. It automatically expands the smaller array to match the larger array, making computations efficient and memory-friendly.

## 🔹 What is Broadcasting?

Broadcasting is the process by which NumPy automatically matches the shapes of arrays during arithmetic operations.

Instead of creating duplicate data, NumPy logically stretches the smaller array to perform element-wise operations.

## ❓ Why Broadcasting?

- Eliminates unnecessary loops.
- Improves performance.
- Saves memory.
- Simplifies array operations.
- Makes mathematical expressions concise.

## 📏 Broadcasting Rules

Two arrays are compatible if:

1. Their dimensions are equal.
2. One of the dimensions is **1**.
3. Missing dimensions are treated as **1**.

```python
import numpy as np

arr = np.array([10, 20, 30])

print(arr + 5)
```

```python
a = np.array([1, 2, 3])

b = np.array([10, 20, 30])

print(a + b)
```

```python
a = np.array([[1],
              [2],
              [3]])

b = np.array([10, 20, 30])

print(a + b)
```

```python
arr = np.array([[1],
                [2],
                [3]])

print(arr * 10)
```

## ❌ Incompatible Shapes

If the array shapes do not satisfy broadcasting rules, NumPy raises a `ValueError`.

```python
a = np.array([1, 2, 3])

b = np.array([1, 2])

print(a + b)
```

ValueError: operands could not be broadcast together
