# 📊 NumPy Statistical Functions

## 🎯 Learning Objectives

- Understand statistical functions in NumPy.
- Calculate measures of central tendency.
- Calculate measures of dispersion.
- Find minimum, maximum, and index positions.
- Perform percentile and quantile analysis.

## 📖 Introduction

NumPy provides built-in statistical functions to analyze numerical data efficiently. These functions help summarize datasets and are widely used in Data Analysis, Exploratory Data Analysis (EDA), and Machine Learning.

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
```

## 🔹 `np.sum()`

Returns the sum of all elements.

```python
print(np.sum(arr))
```

## 🔹 `np.mean()`

Returns the average value of the array.

```python
print(np.mean(arr))
```

## 🔹 `np.median()`

Returns the middle value after sorting the data.

```python
print(np.median(arr))
```

## 🔹 `np.std()`

Measures how much the values deviate from the mean.

```python
print(np.std(arr))
```

## 🔹 `np.var()`

Measures the spread of the data.

```python
print(np.var(arr))
```

## 🔹 `np.min()`

Returns the smallest value.

```python
print(np.min(arr))
```

## 🔹 `np.max()`

Returns the largest value.

```python
print(np.max(arr))
```

## 🔹 `np.argmin()`

Returns the index of the minimum value.

```python
print(np.argmin(arr))
```

## 🔹 `np.argmax()`

Returns the index of the maximum value.

```python
print(np.argmax(arr))
```

## 🔹 `np.percentile()`

Returns the value below which a given percentage of observations fall.

```python
print(np.percentile(arr, 25))
print(np.percentile(arr, 50))
print(np.percentile(arr, 75))
```

## 🔹 `np.quantile()`

Returns values at specified quantiles.

```python
print(np.quantile(arr, 0.25))
print(np.quantile(arr, 0.50))
print(np.quantile(arr, 0.75))
```

## 🔹 `np.average()`

Returns the average of array elements. It also supports weighted averages.

```python
print(np.average(arr))
```

| Function          | Purpose                    |
| ----------------- | -------------------------- |
| `np.sum()`        | Sum of elements            |
| `np.mean()`       | Average                    |
| `np.median()`     | Middle value               |
| `np.std()`        | Standard Deviation         |
| `np.var()`        | Variance                   |
| `np.min()`        | Minimum value              |
| `np.max()`        | Maximum value              |
| `np.argmin()`     | Index of minimum value     |
| `np.argmax()`     | Index of maximum value     |
| `np.percentile()` | Percentile                 |
| `np.quantile()`   | Quantile                   |
| `np.average()`    | Average / Weighted Average |


## 📝 Summary

- NumPy provides powerful statistical functions.
- Mean and median measure the central tendency.
- Variance and standard deviation measure data spread.
- Percentiles and quantiles help understand data distribution.
- These functions are widely used in EDA and Machine Learning.
