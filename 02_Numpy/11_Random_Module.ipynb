# 🎲 NumPy Random Module

## 🎯 Learning Objectives

- Understand random number generation in NumPy.
- Generate random integers and floating-point numbers.
- Create normally distributed random values.
- Randomly select, shuffle, and permute data.
- Ensure reproducible results using random seeds.

## 📖 Introduction

The NumPy Random Module provides functions to generate random numbers for simulations, testing, data sampling, machine learning, and statistical analysis. Setting a random seed ensures reproducible results, which is essential for experiments and model evaluation.

````python
import numpy as np
````

## 🔹 `np.random.seed()`

Initializes the random number generator to produce the same sequence of random numbers every time the program runs.

````python
np.random.seed(42)

print(np.random.rand(3))
````

## 🔹 `np.random.rand()`

Generates random floating-point numbers between **0 and 1**.

````python
print(np.random.rand())

print(np.random.rand(5))

print(np.random.rand(2,3))
````

## 🔹 `np.random.randn()`

Generates random numbers from the standard normal distribution (mean = 0, standard deviation = 1).

````python
print(np.random.randn())

print(np.random.randn(4))

print(np.random.randn(2,3))
````

## 🔹 `np.random.randint()`

Generates random integers within a specified range.

**Syntax**

```python
np.random.randint(low, high, size)
```

````python
print(np.random.randint(1,10))

print(np.random.randint(1,100,5))

print(np.random.randint(0,50,(2,3)))
````

## 🔹 `np.random.choice()`

Randomly selects values from a given array or list.

````python
colors = ["Red","Blue","Green","Yellow"]

print(np.random.choice(colors))

print(np.random.choice(colors,3))
````

## 🔹 `np.random.shuffle()`

Randomly shuffles the elements of an array **in-place**.

````python
arr = np.array([1,2,3,4,5])

np.random.shuffle(arr)

print(arr)
````

## 🔹 `np.random.permutation()`

Returns a shuffled copy of an array without modifying the original array.

````python
arr = np.array([1,2,3,4,5])

print(np.random.permutation(arr))

print(arr)
````

## 🔹 Random Sampling

Randomly select multiple elements from an array.

````python
numbers = np.arange(1,21)

sample = np.random.choice(numbers,5,replace=False)

print(sample)
````

## 🔹 Generate Random Matrix

````python
matrix = np.random.randint(1,100,(3,3))

print(matrix)
````

| Function        | Purpose                     |
| --------------- | --------------------------- |
| `seed()`        | Reproducible random numbers |
| `rand()`        | Random floats (0–1)         |
| `randn()`       | Normal distribution         |
| `randint()`     | Random integers             |
| `choice()`      | Random selection            |
| `shuffle()`     | Shuffle original array      |
| `permutation()` | Shuffled copy               |
| Random Sampling | Select random samples       |


## 📝 Summary

- NumPy provides efficient functions for generating random numbers.
- `seed()` ensures reproducibility.
- `rand()` generates uniformly distributed values.
- `randn()` generates normally distributed values.
- `shuffle()` modifies the original array, while `permutation()` returns a new shuffled array.
- Random sampling is widely used in Machine Learning and Data Analysis.
