# 🧩 Creating Arrays in NumPy

## 🎯 Learning Objectives

- Create NumPy arrays.
- Learn different array creation methods.
- Understand the `dtype` parameter.
- Create 1D, 2D, and 3D arrays.
- Generate special arrays.

## 📖 Introduction

NumPy provides multiple ways to create arrays for different use cases. Arrays are the core data structure in NumPy and are the foundation for numerical computing and machine learning.

## 📦 Creating Arrays from Python Lists

Convert a Python list into a NumPy array using `np.array()`.

```python
import numpy as np 
arr = np.array([10,20,30,40])
print(arr)
```

## 📊 Creating a 2D Array

A 2D array represents rows and columns, similar to a matrix.

```python
import numpy as np
arr = np.array([[10,20,30],
               [40,50,60]])
print(arr)
```

## 🧊 Creating a 3D Array

A 3D array consists of multiple 2D arrays stacked together.

```python
import numpy as np
arr = np.array([
    [[1,2], [3,4],[5,6]]])
print(arr)
```

## 🔢 Specifying Data Type (`dtype`)

Use the `dtype` parameter to define the type of elements stored in the array.

```python
arr = np.array([1,2,3], dtype=float)
print(arr)
```

## 0️⃣ np.zeros()

Creates an array filled with zeros.

```python
arr = np.zeros((2,3))
print(arr)
```

## 1️⃣ np.ones()

Creates an array filled with ones.

```python
arr = np.ones((2,3))
print(arr)
```

## 📄 np.empty()

Creates an array without initializing values.

```python
arr = np.empty((2,3))
print(arr)
```

## 🆔 np.eye()

Creates an identity matrix.

```python
arr = np.eye(2)
print(arr)
```

## 🔢 np.arange()

Creates evenly spaced values within a given range.

```python
arr = np.arange(0, 10, 2)
print(arr)
```

## 📏 np.linspace()

Creates evenly spaced numbers between two limits.

```python
arr = np.linspace(0,1,5)
print(arr)
```

## 🎲 Random Arrays

Generate random numbers using NumPy.

```python
arr = np.random.rand(3,3)
print(arr)
```

```python
arr = np.random.randint(1,100, (3,3))
print(arr)
```
