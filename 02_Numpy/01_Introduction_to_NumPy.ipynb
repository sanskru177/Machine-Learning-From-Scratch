## 📖 Introduction

NumPy (Numerical Python) is the core library for numerical computing in Python. It provides fast, memory-efficient multidimensional arrays and mathematical functions, making it the foundation of Data Science, Machine Learning, and Scientific Computing.

## 🤖 What is NumPy?

NumPy is an open-source Python library used for:

- Numerical Computing
- N-dimensional Arrays (`ndarray`)
- Matrix Operations
- Linear Algebra
- Statistics
- Random Number Generation

It is faster and more memory-efficient than Python lists.

## ❓ Why NumPy?

- Faster computation
- Lower memory usage
- Supports vectorized operations
- Optimized C implementation
- Backbone of Pandas, SciPy, TensorFlow, and Scikit-learn

## ⭐ Key Features

- High-performance arrays
- Multi-dimensional data
- Broadcasting
- Universal Functions (ufuncs)
- Mathematical & Statistical operations
- Linear Algebra support
- Random module

## ⚖️ NumPy vs Python Lists

| Python List              | NumPy Array |
|-------------|            |-------------|
| Slower                   | Faster |
| More Memory              | Less Memory |
| Limited Math Operations  | Vectorized Operations |
| General Purpose          | Numerical Computing |
