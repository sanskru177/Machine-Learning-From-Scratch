# ➕ NumPy Array Operations

## 📖 Introduction

NumPy allows mathematical operations to be performed directly on arrays without using loops. These operations are fast, efficient, and are the foundation of Data Science and Machine Learning.

```python
import numpy as np

arr1 = np.array([10, 20, 30, 40])
arr2 = np.array([1, 2, 3, 4])

print(arr1)
print(arr2)
```

## ➕ Addition (`+`)

```python
print(arr1 + arr2)
```

## ➖ Subtraction (`-`)

```python
print(arr1 - arr2)
```

## ✖️ Multiplication (`*`)

```python
print(arr1 * arr2)
```

## ➗ Division (`/`)

```python
print(arr1 / arr2)
```

## 🔢 Floor Division (`//`)

```python
print(arr1 // arr2)
```

## 📌 Modulus (`%`)

```python
print(arr1 % arr2)
```

## ⚡ Power (`**`)

```python
print(arr2 ** 2)
```

## 🔹 Element-wise Operations

NumPy performs operations element by element.

```python
print(arr1 + arr2)
print(arr1 * arr2)
```

## 🔹 Scalar Operations

Apply operations between an array and a single value.

```python
print(arr1 + 10)
print(arr1 * 2)
print(arr1 / 5)
```

## ⚖️ Comparison Operators

```python
print(arr1 > 20)
print(arr1 < 20)
print(arr1 == 20)
print(arr1 != 20)
print(arr1 >= 20)
print(arr1 <= 20)
```

## 📊 Aggregate Functions

```python
print(np.sum(arr1))
print(np.min(arr1))
print(np.max(arr1))
print(np.mean(arr1))
print(np.prod(arr2))
```

## 🧮 Mathematical Functions

```python
print(np.sqrt(arr1))
print(np.square(arr2))
print(np.abs([-10, 20, -30]))
print(np.round([2.35, 3.75]))
```

## 📝 Summary

- NumPy performs element-wise operations.
- Supports arithmetic, comparison, and scalar operations.
- Aggregate functions summarize array values.
- Mathematical functions simplify numerical computations.
- Operations are faster than Python loops.
