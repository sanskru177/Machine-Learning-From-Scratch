## 📖 Introduction

Decision Tree Regression is a supervised machine learning algorithm used to predict continuous numerical values.

Unlike Linear Regression, it does not assume a linear relationship.

Instead, it divides the feature space into smaller regions using a tree-like structure and predicts the average value of each region.

## ❓ What is Decision Tree Regression?

Decision Tree Regression creates a tree by repeatedly splitting the data into smaller subsets.

Each split is chosen to reduce the prediction error.

Finally, each leaf node stores the average target value of the training samples within that region.

## 🌍 Real-Life Intuition

Imagine a company deciding employee salaries.

Instead of fitting a straight line, the company creates rules like:

If Level ≤ 3 → Salary ≈ ₹50,000

If Level > 3 and ≤ 6 → Salary ≈ ₹120,000

If Level > 6 → Salary ≈ ₹400,000

The Decision Tree learns these rules automatically.

## 🌳 How Does It Work?

Step 1:

Start with the complete dataset.

↓

Step 2:

Try different split points.

↓

Step 3:

Choose the split that minimizes prediction error.

↓

Step 4:

Repeat until stopping criteria are met.

↓

Step 5:

Each leaf predicts the average value.

## 📚 Terminology

Root Node

The first split of the tree.

Internal Node

Intermediate decision points.

Leaf Node

Final prediction.

Depth

Maximum number of splits.

Branch

Connection between nodes.

## 📊 Variance Reduction

Decision Tree Regression selects splits that reduce variance.

Lower variance means samples within a node are more similar.

The algorithm continues splitting until variance cannot be reduced significantly.

## ✅ Advantages

- No linearity assumption
- Handles nonlinear relationships
- Easy to interpret
- No feature scaling required
- Handles outliers better than Linear Regression

## ❌ Disadvantages

- Can overfit easily
- Sensitive to small changes
- Lower generalization if not pruned

## 🌍 Applications

- House Price Prediction
- Salary Prediction
- Demand Forecasting
- Sales Prediction
- Financial Modeling

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeRegressor
```

```python
df = pd.read_csv(r"D:\Sanskruti\sanskruti document\Machine-Learning-From-Scratch\datasets\Position_Salaries.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
X = df[["Level"]]

y = df["Salary"]
```

```python
model = DecisionTreeRegressor(
    random_state=42
)
```

```python
model.fit(X, y)
```

```python
salary = model.predict([[6.5]])

print("Predicted Salary:", salary[0])
```

```python
X_grid = np.arange(
    min(X.values),
    max(X.values),
    0.01
)

X_grid = X_grid.reshape(-1,1)

plt.figure(figsize=(8,5))

plt.scatter(X, y)

plt.plot(
    X_grid,
    model.predict(X_grid),
    color="red"
)

plt.title("Decision Tree Regression")

plt.xlabel("Level")

plt.ylabel("Salary")

plt.show()
```

## ⚙️ Hyperparameters

max_depth
Maximum tree depth.

min_samples_split

Minimum samples required to split.

min_samples_leaf

Minimum samples in a leaf node.

random_state

Controls randomness.

## ⚠️ Overfitting

Decision Trees can memorize the training data.

Signs of Overfitting:

- Very deep tree
- Excellent training accuracy
- Poor testing performance

Reduce overfitting by:

- Limiting max_depth
- Increasing min_samples_leaf
- Pruning

## ✅ Best Practices

- Tune max_depth.
- Use cross-validation.
- Compare with Random Forest.
- Avoid extremely deep trees.

## ⚠️ Common Mistakes

- Allowing unlimited depth.
- Ignoring overfitting.
- Evaluating only on training data.

## 📝 Summary

Decision Tree Regression predicts continuous values by recursively splitting the data into smaller regions. It handles nonlinear relationships well but requires careful tuning to avoid overfitting.
