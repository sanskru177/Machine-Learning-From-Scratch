## 📖 Introduction

Random Forest Regression is an ensemble learning algorithm that combines multiple Decision Trees to make more accurate and stable predictions.

Instead of relying on a single tree, it builds many trees and averages their predictions.

This reduces overfitting and improves generalization.

## ❓ What is Random Forest?

Random Forest is a collection of Decision Trees.

Each tree is trained on a different random subset of the training data.

The final prediction is the average prediction of all trees.

## 🌍 Real-Life Intuition

Imagine you want to estimate the price of a house.

Instead of asking one expert, you ask 100 experts.

Each expert gives a slightly different estimate.

You average all estimates.

The final prediction is usually more reliable than relying on one expert.

## 🌲 How Does Random Forest Work?

Step 1

Create many random datasets using Bootstrap Sampling.

↓

Step 2

Train one Decision Tree on each dataset.

↓

Step 3

Each tree predicts a value.

↓

Step 4

Average all predictions.

↓

Final Prediction

## 📦 Bootstrap Sampling

Bootstrap Sampling means creating multiple datasets by randomly selecting samples from the original dataset **with replacement**.

This means the same data point can appear multiple times in one bootstrap sample, while some data points may not appear at all.

Each Decision Tree is trained on a different bootstrap sample, making the trees diverse.

## 🎲 Feature Randomness

At each split, Random Forest considers only a random subset of the available features instead of all features.

Benefits:

- Reduces correlation between trees.
- Increases diversity.
- Improves overall performance.

## ✅ Advantages

- High accuracy
- Handles nonlinear relationships
- Less overfitting than Decision Trees
- Works well with large datasets
- No feature scaling required

## ❌ Disadvantages

- Slower than Decision Trees
- Larger memory usage
- Less interpretable

## 🌍 Applications

- House Price Prediction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv(r"D:\Sanskruti\sanskruti document\Machine-Learning-From-Scratch\datasets\Position_Salaries.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
X = df[["Level"]]

y = df["Salary"]
```

```python
model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)
```

```python
model.fit(X, y)
```

```python
salary = model.predict([[6.5]])

print("Predicted Salary:", salary[0])
```

```python
X_grid = np.arange(
    min(X.values),
    max(X.values),
    0.01
)

X_grid = X_grid.reshape(-1,1)

plt.figure(figsize=(8,5))

plt.scatter(X, y)

plt.plot(
    X_grid,
    model.predict(X_grid),
    color="red"
)

plt.title("Random Forest Regression")

plt.xlabel("Level")

plt.ylabel("Salary")

plt.show()
```

## ⚙️ Hyperparameters

### n_estimators

Number of Decision Trees.

Default: 100

---

### max_depth

Maximum tree depth.

---

### min_samples_split

Minimum samples required to split.

---

### min_samples_leaf

Minimum samples required in a leaf.

---

### max_features

Number of features considered at each split.

## 📊 Comparison

| Decision Tree | Random Forest |
|--------------|---------------|
| Single tree | Multiple trees |
| High overfitting | Less overfitting |
| Faster | Slightly slower |
| Less accurate | More accurate |
| Easy to interpret | Harder to interpret |

## ⚠️ Overfitting

Random Forest reduces overfitting by averaging predictions from multiple Decision Trees.

Increasing the number of trees generally improves stability, though it also increases computation time.

## ✅ Best Practices

- Use sufficient trees (`n_estimators`).
- Tune `max_depth`.
- Use cross-validation.
- Compare with Gradient Boosting.

## ⚠️ Common Mistakes

- Using very few trees.
- Ignoring hyperparameter tuning.
- Assuming more trees always mean better performance.

## 📝 Summary

Random Forest Regression combines many Decision Trees to produce accurate and stable predictions. By using bootstrap sampling and feature randomness, it reduces overfitting and generally performs better than a single Decision Tree.
