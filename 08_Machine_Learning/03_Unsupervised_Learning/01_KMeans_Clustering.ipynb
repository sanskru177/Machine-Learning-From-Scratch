## 📖 Introduction

K-Means is one of the most popular unsupervised machine learning algorithms.

Unlike supervised learning, K-Means does not require labeled data.

Its objective is to divide data into K clusters such that similar data points belong to the same cluster.

## ❓ What is Unsupervised Learning?

Unsupervised Learning finds hidden patterns in data without target labels.

Examples:

- Customer Segmentation
- Document Clustering
- Image Segmentation
- Recommendation Systems

## 📊 What is Clustering?

Clustering groups similar data points together.

Points inside one cluster are more similar to each other than to points in other clusters.

## 🌍 Real-Life Intuition

Imagine a shopping mall.

Customers have different shopping habits.

Some spend a lot.

Some visit frequently.

Some purchase only occasionally.

K-Means automatically groups similar customers into different clusters.

## ⚙️ Working of K-Means

Step 1

Choose K clusters.

↓

Step 2

Initialize K centroids randomly.

↓

Step 3

Assign every point to the nearest centroid.

↓

Step 4

Update centroid positions.

↓

Step 5

Repeat until centroids stop changing.

## ⭐ Centroid

A centroid is the center of a cluster.

Every data point is assigned to the nearest centroid.

The centroid is updated after each iteration.

## 📊 Within Cluster Sum of Squares (WCSS)

WCSS measures the compactness of clusters.

Lower WCSS indicates tighter and better clusters.

K-Means tries to minimize WCSS.

## 📈 Elbow Method

The Elbow Method helps determine the optimal number of clusters.

Train K-Means with different K values.

Plot K vs WCSS.

Choose the point where the decrease in WCSS becomes less significant (the "elbow").

## ✅ Advantages

- Simple
- Fast
- Easy to implement
- Works well on large datasets

## ❌ Disadvantages

- Need to choose K manually
- Sensitive to outliers
- Sensitive to centroid initialization
- Assumes spherical clusters

## 🌍 Applications

- Customer Segmentation
- Market Basket Analysis
- Image Compression
- Document Clustering
- Recommendation Systems
