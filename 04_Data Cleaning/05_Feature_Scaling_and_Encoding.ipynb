## 📖 Introduction

Machine learning models work with numerical data. Therefore, categorical features must be converted into numerical values through encoding. Additionally, numerical features often have different scales, which may negatively impact model performance.

Feature scaling standardizes numerical features, while encoding transforms categorical features into machine-readable formats.

## ❓ Why Encoding?

Machine learning algorithms cannot directly understand text values such as:

Male
Female

Cherbourg
Southampton

These values must be converted into numerical representations before model training.

## 📚 Common Encoding Techniques

- Label Encoding
- One-Hot Encoding
- Ordinal Encoding

```python
import pandas as pd

from sklearn.preprocessing import (
    LabelEncoder,
    OneHotEncoder,
    StandardScaler,
    MinMaxScaler,
    RobustScaler
)
```

```python
df = pd.read_csv(r"D:\Sanskruti\sanskruti document\Machine-Learning-From-Scratch\datasets\train.csv")

df.head()
```

### Label Encoding

Label Encoding assigns a unique integer to each category.

Example

Male → 1

Female → 0

Best suited for binary or ordinal features.

```python
le = LabelEncoder()

df["Sex"] = le.fit_transform(df["Sex"])

df[["Sex"]].head()
```

### One-Hot Encoding

Creates a separate column for each category.

Example

Embarked

S

C

Q

becomes

Embarked_C

Embarked_Q

Embarked_S

This avoids assigning numerical order where none exists.

```python
encoded = pd.get_dummies(df["Embarked"], prefix="Embarked")

encoded.head()
```

```python
df = pd.concat([df, encoded], axis=1)

df.head()
```

## 📏 Why Feature Scaling?

Different features may have different ranges.

Example

Age → 5 to 80

Fare → 0 to 512

Large-scale features may dominate smaller ones during model training.

Scaling ensures all features contribute fairly.

### StandardScaler

Formula

z = (x − μ) / σ

Produces

Mean ≈ 0

Standard Deviation ≈ 1

Best used for algorithms such as Logistic Regression, SVM, and PCA.

```python
scaler = StandardScaler()

df["Age_scaled"] = scaler.fit_transform(df[["Age"]].fillna(df["Age"].median()))

df[["Age","Age_scaled"]].head()
```

### MinMaxScaler

Transforms values between 0 and 1.

Useful for Neural Networks and KNN.

```python
minmax = MinMaxScaler()

df["Fare_scaled"] = minmax.fit_transform(df[["Fare"]])

df[["Fare","Fare_scaled"]].head()
```

### RobustScaler

Uses Median and IQR instead of Mean and Standard Deviation.

Best suited for datasets containing outliers.

```python
robust = RobustScaler()

df["Fare_robust"] = robust.fit_transform(df[["Fare"]])

df[["Fare","Fare_robust"]].head()
```

```python
df[[
    "Age",
    "Age_scaled",
    "Fare",
    "Fare_scaled",
    "Fare_robust"
]].head()
```

## 🔍 Observations

- Label Encoding converted the Sex column into numerical values.
- One-Hot Encoding created separate columns for each Embarked category.
- StandardScaler standardized the Age feature.
- MinMaxScaler transformed Fare into the range [0,1].
- RobustScaler handled the Fare feature while reducing the influence of outliers.
