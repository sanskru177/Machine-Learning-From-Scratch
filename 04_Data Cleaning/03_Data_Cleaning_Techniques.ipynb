## 📖 Introduction

Data cleaning is the process of improving the quality of a dataset by identifying and correcting errors, inconsistencies, and unwanted information.

A clean dataset leads to more reliable analysis and better machine learning models. Data scientists spend a significant portion of their time cleaning and preparing data before building models.

## 🔁 Removing Duplicate Data

Duplicate records are identical rows that appear more than once in a dataset.

### Why Remove Duplicates?

- Reduces data redundancy.
- Prevents biased analysis.
- Improves model accuracy.
- Saves storage and processing time.

```python
import pandas as pd
df = pd.read_csv(r"D:\Sanskruti\sanskruti document\Machine-Learning-From-Scratch\datasets\train.csv")

df.duplicated().sum()
```

```python
df = df.drop_duplicates()

df.shape
```

## 🔄 Data Type Conversion

Sometimes columns are stored with incorrect data types.

Converting them to the correct type improves memory usage and enables proper analysis.

```python
df.dtypes
```

```python
df["Age"] = df["Age"].astype("float")
```

```python
df.dtypes
```

## 📝 Renaming Columns

Meaningful column names improve code readability and make datasets easier to understand.

```python
df.rename(columns={"PassengerId":"Passenger_ID"}, inplace=True)

df.head()
```

## ✨ String Cleaning

Text data often contains unwanted spaces, inconsistent letter cases, or special characters.

Cleaning text ensures consistency across the dataset.

```python
df["Name"] = df["Name"].str.strip()
```

```python
df["Name"] = df["Name"].str.lower()
```

```python
df["Name"] = df["Name"].str.upper()
```

```python
df["Name"] = df["Name"].str.replace(",", "", regex=False)
```

## 🗑 Removing Unnecessary Columns

Some columns provide little value for analysis or machine learning.

Removing irrelevant features reduces complexity and improves efficiency.

```python
df.drop(columns=["Ticket"], inplace=True)

df.head()
```

## 🔍 Observations

- No duplicate rows are present in the Titanic dataset.
- Data types should always be verified before analysis.
- Renaming columns improves readability.
- String cleaning removes inconsistencies.
- Removing unnecessary columns simplifies the dataset.

## ✅ Best Practices

- Always inspect the dataset before cleaning.
- Keep meaningful column names.
- Remove duplicate rows before analysis.
- Clean string data for consistency.
- Drop only columns that do not contribute to analysis.
