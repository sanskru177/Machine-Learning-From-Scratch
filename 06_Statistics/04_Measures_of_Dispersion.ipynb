## 📖 Introduction

Measures of Central Tendency (Mean, Median, Mode) tell us the center of the data, but they do not tell us how spread out the data is.

Measures of Dispersion describe the variability or spread of observations around the center.

Understanding dispersion helps us compare datasets and detect unusual values (outliers).

## ❓ What is Dispersion?

Dispersion refers to how much the data values differ from each other and from the central value.

Example:

Dataset A:
20, 21, 22, 23, 24

Dataset B:
5, 15, 22, 30, 40

Both datasets have a similar average, but Dataset B is much more spread out.

## 📚 Types of Dispersion

- Range
- Variance
- Standard Deviation
- Quartiles
- Percentiles
- Interquartile Range (IQR)

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"D:\Sanskruti\sanskruti document\Machine-Learning-From-Scratch\datasets\train.csv")
```

## 📌 Range

Range is the difference between the maximum and minimum values.

### Formula

Range = Maximum − Minimum

Example:

Marks:
50, 60, 70, 80, 90

Range = 90 − 50 = 40

A larger range indicates greater variability.

```python
age_range = df["Age"].max() - df["Age"].min()

print(age_range)
```

## 📌 Variance

Variance measures how far each value is from the mean on average.

A low variance means the data points are close to the mean.

A high variance means the data points are more spread out.

Variance is expressed in squared units.

```python
df["Age"].var()
```

## 📌 Standard Deviation

Standard Deviation is the square root of the variance.

Unlike variance, it is expressed in the same unit as the original data, making it easier to interpret.

Interpretation:

- Low Standard Deviation → Data is close to the mean.
- High Standard Deviation → Data is widely spread.

```python
df["Age"].std()
```

## 📌 Difference Between Variance and Standard Deviation

| Variance             | Standard Deviation      |
| -------------------- | ----------------------- |
| Square of deviations | Square root of variance |
| Squared units        | Original units          |
| Harder to interpret  | Easier to interpret     |


## 📌 Quartiles

Quartiles divide the dataset into four equal parts.

- Q1 (25th Percentile)
- Q2 (50th Percentile / Median)
- Q3 (75th Percentile)

Quartiles help understand the distribution of data.

```python
df["Age"].quantile([0.25, 0.50, 0.75])
```

## 📌 Percentiles

Percentiles divide data into 100 equal parts.

Examples:

- 25th Percentile
- 50th Percentile
- 90th Percentile
- 95th Percentile

Percentiles are widely used in exams and performance analysis.

```python
df["Age"].quantile(0.90)
```

## 📌 Interquartile Range (IQR)

The Interquartile Range measures the spread of the middle 50% of the data.

### Formula

IQR = Q3 − Q1

IQR is less affected by outliers than the range.

```python
Q1 = df["Age"].quantile(0.25)

Q3 = df["Age"].quantile(0.75)

IQR = Q3 - Q1

print(IQR)
```

## 📌 Detecting Outliers Using IQR
Lower and Upper Limits

```python
lower_limit = Q1 - 1.5 * IQR

upper_limit = Q3 + 1.5 * IQR

print(lower_limit)

print(upper_limit)
```

Find Outliers

```python
outliers = df[(df["Age"] < lower_limit) | (df["Age"] > upper_limit)]

outliers.head()
```

## 🔍 Observations

- Range provides a quick measure of spread.
- Variance measures average squared deviation from the mean.
- Standard Deviation is easier to interpret because it uses the original units.
- IQR is effective for detecting outliers.
- Quartiles and Percentiles help summarize the distribution of data.

## ✅ Best Practices

- Use Standard Deviation instead of Variance for interpretation.
- Use IQR to detect outliers.
- Compare dispersion along with measures of central tendency.
- Visualize the data before making conclusions.

## 🌍 Real-World Scenario

A school compares the marks of two classes.

- Class A has an average score of 75 with a low standard deviation, meaning most students scored close to 75.
- Class B also has an average score of 75 but a much higher standard deviation, meaning students' scores varied widely.

Although the averages are the same, the consistency of performance differs significantly.
