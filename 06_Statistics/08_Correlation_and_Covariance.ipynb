## 📖 Introduction

In Data Analysis and Machine Learning, we often want to know whether two variables are related.

For example:

- Does Age affect Fare?
- Does Study Time affect Exam Marks?
- Does Experience affect Salary?

Correlation and Covariance help us answer these questions.

## ❓ Covariance

Covariance measures how two variables change together.

Interpretation:

- Positive Covariance → Both variables increase together.
- Negative Covariance → One increases while the other decreases.
- Zero Covariance → No linear relationship.

Covariance tells the **direction** of the relationship but not its strength.

## ❓ Correlation

Correlation measures both the **strength** and **direction** of the linear relationship between two variables.

The correlation coefficient always lies between:

-1 ≤ r ≤ 1

Where:

- +1 → Perfect Positive Correlation
- 0 → No Linear Correlation
- -1 → Perfect Negative Correlation

## 📚 Types of Correlation

### Positive Correlation

As one variable increases, the other also increases.

Example:

Experience ↑ → Salary ↑

---

### Negative Correlation

As one variable increases, the other decreases.

Example:

Price ↑ → Demand ↓

---

### Zero Correlation

No relationship between variables.

Example:

Shoe Size and IQ (generally unrelated).

## 📌 Covariance vs Correlation

| Covariance           | Correlation                  |
| -------------------- | ---------------------------- |
| Shows direction only | Shows direction and strength |
| Can take any value   | Always between -1 and 1      |
| Depends on units     | Unit-free                    |
| Harder to interpret  | Easier to interpret          |


## 📌 Pearson Correlation

Pearson Correlation measures the linear relationship between two numerical variables.

Use when:

- Data is numerical.
- Relationship is approximately linear.
- Data is approximately normally distributed.

## 📌 Spearman Correlation

Spearman Correlation measures the relationship between ranked variables.

Use when:

- Data is ordinal.
- Data is not normally distributed.
- Relationship is monotonic but not necessarily linear.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Sanskruti\sanskruti document\Machine-Learning-From-Scratch\datasets\train.csv")
```

```python
numerical_df = df.select_dtypes(include=["int64", "float64"])

numerical_df.head()
```

```python
correlation_matrix = numerical_df.corr()

correlation_matrix
```

```python
df["Age"].corr(df["Fare"])
```

📌 Covariance Between Age and Fare

```python
df["Age"].cov(df["Fare"])
```

📌 Heatmap (Using Matplotlib)

```python
plt.figure(figsize=(8,6))

plt.imshow(correlation_matrix, cmap="coolwarm")

plt.colorbar()

plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)

plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)

plt.title("Correlation Matrix")

plt.show()
```

📌 Scatter Plot

```python
plt.figure(figsize=(6,4))

plt.scatter(df["Age"], df["Fare"])

plt.xlabel("Age")

plt.ylabel("Fare")

plt.title("Age vs Fare")

plt.show()
```

## 🔍 Observations

- Correlation measures both strength and direction.
- Covariance measures only direction.
- Correlation values close to ±1 indicate strong relationships.
- Correlation does not imply causation.

## ⚠️ Important Note

Two variables may be highly correlated without one causing the other.

Example:

Ice cream sales and drowning incidents both increase during summer, but buying ice cream does not cause drowning. The hidden factor is hot weather.

## 🌍 Applications

Correlation is used in:

- Feature Selection
- Stock Market Analysis
- Healthcare
- Marketing
- Finance
- Recommendation Systems
- Machine Learning

## ✅ Best Practices

- Use Correlation for numerical variables.
- Remove highly correlated features to reduce multicollinearity.
- Always visualize relationships using scatter plots.
- Remember that correlation alone cannot establish cause-and-effect.
