## 📖 Introduction

Understanding data types also helps in selecting the correct visualization techniques and preprocessing methods.

## ❓ What is Data?

Data is a collection of facts, measurements, observations, or information.

Examples:

- Age = 25
- Salary = ₹50,000
- Gender = Female
- City = Mumbai
- Temperature = 36.5°C

These values help us analyze patterns and make decisions.

## 📚 Main Types of Data

There are two main types of data:

1. Qualitative (Categorical)
2. Quantitative (Numerical)

## 📝 Qualitative Data

Qualitative data represents qualities or categories rather than numbers.

Examples:

- Gender
- City
- Blood Group
- Department
- Color

Characteristics:

- Cannot perform mathematical operations.
- Used for grouping or classification.

## 🔢 Quantitative Data

Quantitative data represents measurable quantities.

Examples:

- Age
- Salary
- Height
- Weight
- Marks

Characteristics:

- Mathematical calculations can be performed.
- Used for statistical analysis.

## 📚 Quantitative Data Types

### 1. Discrete Data

Discrete data consists of countable values.

Examples:

- Number of Students
- Number of Cars
- Number of Employees

Examples:

0,1,2,3,4...

---

### 2. Continuous Data

Continuous data can take any value within a range.

Examples:

- Height
- Weight
- Temperature
- Salary
- Age

Examples:

165.5 cm

55.8 kg

36.7°C

## 📚 Categorical Data Types

### Nominal Data

Categories without any order.

Examples:

- Gender
- City
- Blood Group
- Nationality

---

### Ordinal Data

Categories with a meaningful order.

Examples:

- Education Level
- Customer Satisfaction
- Movie Rating

Example:

Poor

Average

Good

Excellent

## 📏 Measurement Scales

### Nominal

Categories only.

Example:

Red, Blue, Green

---

### Ordinal

Ordered categories.

Example:

Low

Medium

High

---

### Interval

Equal intervals but no true zero.

Example:

Temperature in Celsius.

0°C does not mean "no temperature."

---

### Ratio

Equal intervals and a true zero.

Examples:

Weight

Height

Age

Salary

0 means absence of the quantity.
