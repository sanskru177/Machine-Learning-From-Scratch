## 📐 Formula

Probability(Event)

= Number of Favorable Outcomes

/

Total Number of Possible Outcomes

Example:

Rolling a die

Probability of getting 4

= 1 / 6

## 📦 Sample Space

A Sample Space is the set of all possible outcomes.

Example:

Coin Toss

S = {H, T}

Dice Roll

S = {1,2,3,4,5,6}

## 🎯 Event

An Event is one or more outcomes from the sample space.

Example:

Rolling an even number

Event

E = {2,4,6}

## 📚 Types of Events

### Simple Event

Contains only one outcome.

Example:

Rolling a 5.

---

### Compound Event

Contains more than one outcome.

Example:

Rolling an even number.

---

### Certain Event

Probability = 1

Example:

Getting a number between 1 and 6 on a standard die.

---

### Impossible Event

Probability = 0

Example:

Rolling a 7 on a standard die.

## 🔗 Independent Events

Two events are independent if one event does not affect the other.

Example:

- Tossing a coin.
- Rolling a die.

The outcome of the coin toss does not affect the die roll.

## 🔗 Dependent Events

Two events are dependent if one event affects the probability of the other.

Example:

Drawing two cards without replacement.

The first card changes the probability of the second card.

## 📌 Conditional Probability

Conditional Probability is the probability of an event occurring given that another event has already occurred.

Formula:

P(A|B)

=

P(A ∩ B)

/

P(B)

Example:

Probability that a passenger survived given that the passenger is female.

## 📌 Bayes' Theorem

Bayes' Theorem updates the probability of an event using new information.

Formula:

P(A|B)

=

P(B|A) × P(A)

/

P(B)

Applications:

- Spam Detection
- Medical Diagnosis
- Fraud Detection
- Naive Bayes Classifier
