## 📖 Introduction

A Normal Distribution is one of the most important probability distributions in statistics.

It is also known as the Gaussian Distribution or Bell Curve because of its bell-shaped appearance.

Many real-world variables such as height, weight, IQ scores, and exam marks approximately follow a normal distribution.

## ❓ What is a Normal Distribution?

A Normal Distribution is a continuous probability distribution where:

- Most observations are concentrated around the mean.
- Fewer observations occur as we move away from the mean.
- The distribution is symmetric.

Example:

Student marks in a large university often follow a normal distribution.

## 📚 Properties

- Bell-shaped curve.
- Symmetrical around the mean.
- Mean = Median = Mode.
- Total area under the curve = 1 (100% probability).
- Tails extend infinitely but never touch the x-axis.

## 🔔 Bell Curve

The graph of a normal distribution is called a Bell Curve because of its shape.

- Highest point → Mean
- Left and right sides are mirror images.
- Most values lie near the center.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Sanskruti\sanskruti document\Machine-Learning-From-Scratch\datasets\train.csv")
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Age"].dropna(), bins=20)

plt.title("Age Distribution")

plt.xlabel("Age")

plt.ylabel("Frequency")

plt.show()
```

Check whether the Age distribution looks approximately normal or skewed.

```python
print("Mean:", df["Age"].mean())

print("Median:", df["Age"].median())

print("Mode:", df["Age"].mode()[0])
```

Interpretation

If Mean ≈ Median ≈ Mode, the data may be approximately normally distributed.

## 📌 Standard Normal Distribution

A Standard Normal Distribution has:

- Mean = 0
- Standard Deviation = 1

It is used to calculate Z-Scores.

## 📏 Z-Score

A Z-Score tells us how many standard deviations a value is from the mean.

Formula:

Z = (X − Mean) / Standard Deviation

Interpretation:

- Z = 0 → At the mean.
- Z > 0 → Above the mean.
- Z < 0 → Below the mean.

```python
mean = df["Age"].mean()

std = df["Age"].std()

df["Age_Z"] = (df["Age"] - mean) / std

df[["Age", "Age_Z"]].head()
```

## 📚 Empirical Rule

For a Normal Distribution:

- About 68% of values lie within ±1 Standard Deviation.
- About 95% of values lie within ±2 Standard Deviations.
- About 99.7% of values lie within ±3 Standard Deviations.

## 📌 Detect Outliers Using Z-Score

```python
outliers = df[df["Age_Z"].abs() > 3]

outliers.head()
```

Interpretation

Values with |Z| > 3 are often considered potential outliers.

## 🔍 Observations

- Normal Distribution is symmetric.
- Mean, Median, and Mode are approximately equal.
- Most observations lie close to the mean.
- Z-Scores help compare values on a common scale.
- Large absolute Z-Scores may indicate outliers.

## ✅ Best Practices

- Check the distribution before applying statistical techniques.
- Use histograms to inspect the shape of the data.
- Use Z-Scores for approximately normal numerical data.
- Be cautious when interpreting skewed distributions.
