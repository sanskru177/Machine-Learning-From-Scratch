## 📖 Introduction

Measures of Central Tendency describe the central or typical value of a dataset.

Instead of looking at every observation individually, these measures summarize the data using a single representative value.

The three most common measures are:

- Mean
- Median
- Mode

## ❓ What is Central Tendency?

Central Tendency is the value around which most observations are concentrated.

Example:

Marks:

70, 75, 80, 85, 90

The center of these values is around **80**.

## 📚 Types

1. Mean
2. Median
3. Mode

## 📌 Mean

Mean is the sum of all observations divided by the total number of observations.

### Formula

Mean = (Sum of all observations) / (Number of observations)

Example:

Marks:

70, 75, 80, 85, 90

Mean = (70 + 75 + 80 + 85 + 90) / 5 = 80

```python
import pandas as pd

df = pd.read_csv(r"D:\Sanskruti\sanskruti document\Machine-Learning-From-Scratch\datasets\train.csv")

df["Age"].mean()
```

## 📌 Median

Median is the middle value after arranging the data in ascending order.

If the number of observations is odd:
→ Middle value

If even:
→ Average of the two middle values

Example:

5, 8, 10, 12, 15

Median = 10

Example:

5, 8, 10, 12

Median = (8 + 10) / 2 = 9

```python
df["Age"].median()
```

## 📌 Mode

Mode is the value that appears most frequently in a dataset.

Example:

2,2,3,4,4,4,5

Mode = 4

A dataset can have:

- One Mode (Unimodal)
- Two Modes (Bimodal)
- Multiple Modes (Multimodal)
- No Mode (all values occur once)

```python
df["Embarked"].mode()
```

## 📌 Arithmetic Mean

The Arithmetic Mean is the standard average that we commonly use.

Formula:

Arithmetic Mean = Σx / n

It works well when all observations have equal importance.

## 📌 Weighted Mean

Weighted Mean assigns different importance (weights) to observations.

### Formula

Weighted Mean = Σ(wx) / Σw

Example:

Subject | Marks | Credits

Math | 80 | 4

Science | 90 | 3

English | 70 | 2

The weighted average gives more importance to subjects with higher credits.

```python
marks = [80,90,70]

weights = [4,3,2]

weighted_mean = sum(m*w for m,w in zip(marks,weights))/sum(weights)

weighted_mean
```

## 📌 Geometric Mean

The Geometric Mean is useful for growth rates and percentage changes.

Formula:

GM = (x₁ × x₂ × ... × xₙ)^(1/n)

Example:

Growth Rates:

10%

20%

30%

The Geometric Mean provides a more accurate average growth rate than the Arithmetic Mean.

```python
from statistics import geometric_mean

numbers = [2,4,8]

geometric_mean(numbers)
```

## 📌 Harmonic Mean

The Harmonic Mean is useful when averaging rates.

Example:

Average Speed

Average Price per Unit

Formula:

HM = n / Σ(1/x)

```python
from statistics import harmonic_mean

numbers = [60,80]

harmonic_mean(numbers)
```

## 📌 Mean vs Median vs Mode

| Measure | Best Used For            |
| ------- | ------------------------ |
| Mean    | Symmetric numerical data |
| Median  | Data with outliers       |
| Mode    | Categorical data         |


```python
df["Age"].mean()
```

```python
df["Age"].median()
```

```python
df["Embarked"].mode()
```

```python
print("Mean:",df["Age"].mean())

print("Median:",df["Age"].median())

print("Mode:",df["Embarked"].mode()[0])
```

## 🔍 Observations

- Mean uses every observation.
- Median is resistant to outliers.
- Mode is suitable for categorical variables.
- Different datasets may require different measures of central tendency.

## ✅ Best Practices

- Use Mean for normally distributed numerical data.
- Use Median when the data contains outliers or is skewed.
- Use Mode for categorical variables.
- Always visualize the data before choosing a measure of central tendency.

## 🌍 Real-World Scenario

A real estate company analyzes house prices in a city. A few luxury houses worth several crores create extreme outliers, making the average (mean) price much higher than most homes. To represent the typical house price, analysts use the **median** instead of the mean because it is less affected by outliers.
