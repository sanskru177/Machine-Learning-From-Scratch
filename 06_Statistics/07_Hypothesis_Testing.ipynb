## 📖 Introduction

Hypothesis Testing is a statistical method used to determine whether there is enough evidence in a sample to support a particular claim about a population.

Instead of making assumptions, we use sample data and statistical tests to make decisions.

## ❓ What is a Hypothesis?

A hypothesis is a statement or assumption about a population that we want to test.

Example:

"The average age of Titanic passengers is 30 years."

We collect sample data to determine whether this statement is likely to be true.

## 📚 Types of Hypotheses

### Null Hypothesis (H₀)

The Null Hypothesis states that there is **no significant difference** or **no effect**.

Example:

H₀: The average passenger age is 30 years.

---

### Alternative Hypothesis (H₁)

The Alternative Hypothesis states that there **is a significant difference** or **an effect**.

Example:

H₁: The average passenger age is not 30 years.

## 📋 Steps

1. State the Null Hypothesis (H₀).
2. State the Alternative Hypothesis (H₁).
3. Choose the significance level (α).
4. Calculate the test statistic.
5. Find the P-value.
6. Compare the P-value with α.
7. Make a decision.

## 📌 Significance Level

The significance level is the probability of rejecting the Null Hypothesis when it is actually true.

Common values:

- 0.05 (5%)
- 0.01 (1%)

If:

P-value < α

Reject H₀

Otherwise,

Fail to Reject H₀

## 📌 What is a P-Value?

The P-value measures how likely the observed sample data would occur if the Null Hypothesis were true.

Interpretation:

P-value = 0.03

If α = 0.05

0.03 < 0.05

Reject the Null Hypothesis.

## ❌ Type I Error

Rejecting the Null Hypothesis when it is actually true.

Also called:

False Positive

Example:

A healthy patient is diagnosed as having a disease.

## ❌ Type II Error

Failing to reject the Null Hypothesis when it is actually false.

Also called:

False Negative

Example:

A patient with a disease is diagnosed as healthy.

## 📌 One-tailed vs Two-tailed Test

| One-tailed           | Two-tailed             |
| -------------------- | ---------------------- |
| Checks one direction | Checks both directions |
| Example: Mean > 30   | Example: Mean ≠ 30     |


## 📌 Z-Test vs T-Test

| Z-Test                               | T-Test                                |
| ------------------------------------ | ------------------------------------- |
| Large sample size (typically n ≥ 30) | Small sample size (typically n < 30)  |
| Population standard deviation known  | Population standard deviation unknown |
| Uses Normal Distribution             | Uses Student's t-distribution         |


```python
import pandas as pd
from scipy.stats import ttest_1samp
```

```python
df = pd.read_csv(r"D:\Sanskruti\sanskruti document\Machine-Learning-From-Scratch\datasets\train.csv")
```

📌 Example Hypothesis

Suppose we want to test:

H₀: Average passenger age = 30

H₁: Average passenger age ≠ 30

## 📌 Perform One-Sample T-Test

```python
ages = df["Age"].dropna()

t_statistic, p_value = ttest_1samp(ages, popmean=30)

print("T-Statistic:", t_statistic)
print("P-Value:", p_value)
```

📌 Decision

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```

## 🔍 Interpretation

If the P-value is less than 0.05, there is enough statistical evidence to conclude that the average passenger age is significantly different from 30 years.

If the P-value is greater than or equal to 0.05, there is not enough evidence to reject the Null Hypothesis.
