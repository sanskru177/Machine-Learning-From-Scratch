## 📖 Introduction

Statistics is the science of collecting, organizing, analyzing, interpreting, and presenting data.

It helps us transform raw data into meaningful information, enabling better decision-making in business, healthcare, finance, sports, and machine learning.

In Data Science, statistics is one of the most important foundations because every machine learning algorithm relies on statistical concepts.

## ❓ What is Statistics?

Statistics is a branch of mathematics that deals with data.

It involves:

- Collecting data
- Organizing data
- Summarizing data
- Analyzing data
- Drawing conclusions

Statistics helps answer questions such as:

- What is the average salary of employees?
- Which product sells the most?
- How many customers are likely to leave a service?
- Which factors influence house prices?

## 💡 Why is Statistics Important?

Statistics helps us:

- Summarize large datasets.
- Identify trends and patterns.
- Make predictions.
- Compare groups.
- Support business decisions.
- Build machine learning models.
- Evaluate model performance.

## 📚 Types of Statistics

### 1. Descriptive Statistics

Descriptive Statistics summarizes and describes the characteristics of a dataset.

Examples:

- Mean
- Median
- Mode
- Standard Deviation
- Variance
- Range

Example:
The average marks of a class are 78.

---

### 2. Inferential Statistics

Inferential Statistics uses sample data to make conclusions or predictions about an entire population.

Examples:

- Hypothesis Testing
- Confidence Intervals
- Regression Analysis
- ANOVA

Example:
Based on a survey of 500 voters, predict the election result for the whole country.

 ## 📌 Descriptive vs Inferential Statistics

| Feature | Descriptive Statistics | Inferential Statistics                 |
| ------- | ---------------------- | -------------------------------------- |
| Purpose | Summarize data         | Make predictions                       |
| Uses    | Mean, Median, Charts   | Hypothesis Tests, Confidence Intervals |
| Data    | Entire dataset         | Sample data                            |
| Output  | Description            | Conclusions about population           |


## 📌 Common Statistical Terms

## 📖 Important Terms

### Population

The complete collection of all observations.

Example:
All students in a university.

---

### Sample

A subset of the population.

Example:
200 students selected from the university.

---

### Variable

A characteristic measured for each observation.

Examples:

- Age
- Height
- Salary
- Marks

---

### Observation

A single record in a dataset.

Example:

One passenger in the Titanic dataset.
